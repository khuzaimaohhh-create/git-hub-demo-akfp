```python
import cv2

# Read image
img = cv2.imread("Screenshot ... 204742.png")

# Get image size
h, w = img.shape[:2]

# Crop into four parts
top_left = img[:h//2, :w//2]
top_right = img[:h//2, w//2:]
bottom_left = img[h//2:, :w//2]
bottom_right = img[h//2:, w//2:]

# Show images
cv2.imshow("Top Left", top_left)
cv2.imshow("Top Right", top_right)
cv2.imshow("Bottom Left", bottom_left)
cv2.imshow("Bottom Right", bottom_right)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# Read the image
img = cv2.imread("")   

# Check if image loaded
if img is None:
    print("Error: Image not found!")
else:
    # Get image height and width
    h, w = img.shape[:2]

    # Crop into four parts
    top_left = img[0:h//2, 0:w//2]
    top_right = img[0:h//2, w//2:w]
    bottom_left = img[h//2:h, 0:w//2]
    bottom_right = img[h//2:h, w//2:w]

    # Display images
    cv2.imshow("Top Left", top_left)
    cv2.imshow("Top Right", top_right)
    cv2.imshow("Bottom Left", bottom_left)
    cv2.imshow("Bottom Right", bottom_right)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
